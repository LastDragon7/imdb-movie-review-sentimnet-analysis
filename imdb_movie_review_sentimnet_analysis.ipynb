```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/content/IMDB Dataset.csv')
```

```python
dataset.head()
```

```python
dataset['sentiment'] = dataset['sentiment'].replace({'positive':1, 'negative':0})
```


1.   Remove Html tags
2.   remove special character
3. remove stop words
4. convert to lowercase
5. stemming


```python
dataset.head()
```

 **1. Removing HTML**

```python
import re
def remove_html(text):
  clean = re.compile('<.*?>')
  return re.sub(clean, '', text)
```

```python
dataset['review'] = dataset['review'].apply(remove_html)
```

**2. Removing special character**

```python
def remove_special(text):
  ans = ''
  for i in text:
    if i.isalnum():
      ans = ans + i
    else:
      ans = ans + ' '
  return ans
```

```python
dataset['review'] = dataset['review'].apply(remove_special)
dataset.head()
```

```python
dataset['review'] = dataset['review'].apply(lambda x: x.lower())
dataset.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stopwords = stopwords.words('english')

def remove_stopwords(text):
  ans = ''
  for i in text.split():
    if i not in stopwords:
      ans = ans + i + ' '
  return ans
```

```python
dataset['review'] = dataset['review'].apply(remove_stopwords)
dataset.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stemming(text):
  ans = ''
  for i in text.split():
    ans = ans + ps.stem(i) + ' '
  return ans
```

```python
dataset['review'] = dataset['review'].apply(stemming)
dataset.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer( max_features=1000)
```

```python
dataset.shape
```

```python
x = cv.fit_transform(dataset['review']).toarray()
y = dataset['sentiment'].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model1 = GaussianNB()
model2 = MultinomialNB()
model3 = BernoulliNB()
model4 = LogisticRegression()
```

```python
model1.fit(x_train, y_train)
prediction = model1.predict(x_test)
accuracy_score(y_test, prediction)
```

```python
model2.fit(x_train, y_train)
prediction = model2.predict(x_test)
accuracy_score(y_test, prediction)
```

```python
model3.fit(x_train, y_train)
prediction = model3.predict(x_test)
accuracy_score(y_test, prediction)
```

```python
model4.fit(x_train, y_train)
prediction = model4.predict(x_test)
accuracy_score(y_test, prediction)
```
